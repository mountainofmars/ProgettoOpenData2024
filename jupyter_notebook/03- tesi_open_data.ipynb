# 2. ELABORAZIONE DATI

## - PULIZIA


1. Per tutti i dataset viene utilizzato l'encoding UTF-8

2. Lo strumento utilizzato per effettuare tutte le operazioni sui dataset è il pacchetto PANDAS disponibile per Python. 

3. Tutti i titoli delle etichette di ogni dataframe vengono sostituite con la loro versioni in UPPERCASE.

4. Nelle colonne in cui sono presenti date, ne è stata migliorata la formattazione per aumentarne la leggibilità. 

5. In alcuni dataset erano presenti degli errori di ortografia (es: in E-MAIL, alcuni indirizzi erano scritti come xyz@gmail,com) che sono stati corretti tramite regex e la funzione replace() di PANDAS. 

6. Alcune celle dei dataset riguardanti le strutture presentavano "NaN" come valori. </br> Per rendere possibile l'interlinking però è stato necessario rimpiazzare tutti i valori "Nan" con dei valori "dummy" che rappresentassero la mancanza di informazione. </br> ES: nelle celle che dovevano avere string, NaN è stato rimpiazziato con "-"


## ARRICCHIMENTO

- Nel dataset riguardante le **Strutture di ricovero** e le **strutture sanitarie**, per quanto riguarda l'ubicazione, erano presenti informazioni solo riguardanti la regione, il codice regione, indirizzo ed il codice istat del comune. Sono quindi state aggiunte altre due colonne riguardanti le informazioni "DENOMINAZIONE COMUNE" e "PROVINCIA".

```python
import pandas as pd

    #quando apre il csv, considera tutti i dati come "string"
def deletecol(listcol, dfpath, separator):
    df = pd.read_csv(dfpath, dtype=str, sep=separator)
    df.drop(columns= listcol, inplace=True)
    return df


def deletecol_df(listcol, df):
    df.drop(columns= listcol, inplace=True)
    return df
```

## INTERLINK CON DBPEDIA

Per effettuare l'interlink con dbpedia viene usata l'apposita funzione dbpediaInterlink(),
che viene richiamata ogni volta che viene aggiunto un nuovo comune al grafo RDF.
La funzione prende come argomenti “res_com” che indica l’URI del un singolo comune e
“nome_com” che sarebbe la Denominazione del comune.
Per eseguire le queries, la funzione sfrutta la classe **SPARQLWrapper2** contenuta nel pacchetto SPARQLWrapper.
È stato utilizzata questa classe perché ha la particolarità di convertire automaticamente i risultati nel formato JSON. </br> ALGORITMO:

1.	Viene settato lo SPARQL endpoint su cui si andranno ad eseguire le queries.

2.	Viene memorizzata all’interno di una variabile, la query generale che verrà eseguita. In particolare, viene memorizzata come “f-string”, che è un particolare tipo di stringa (formattata) che permette per l’appunto di formattare ed interpolare la stringa a proprio piacimento. È stata usata questa variazione piuttosto che il formato stock di stringa, perché per effettuare l’interlink per ogni comune dovrebbe esserci una query sempre differente; in questo modo viene infatti passato il comune come variabile all’interno della query, permettendo di avere un’unica query flessibile.

3.	STRUTTURA QUERY: vengono estratti i comuni della sicilia e le loro etichette, e l’interlink avviene per un comune alla volta, viene estratto solo il comune passato dalla variabile “nome_com” (con FILTER) , ovviamente in italiano. 

4.	sparql.setQuery(myquery) : viene impostata la query che dovrà essere eseguita.

5.	Viene eseguita la query ed i risultati (detti bindings), vengono memorizzati sotto forma di lista nella variabile “risultati”.

6.	Se c’è almeno un risultato viene effettuato l’interlink aggiungendo al grafo l’URI del risultato estratto dalla query.

7. Il dataset sui comuni comprendeva inizialmente tutti i comuni italiani: il file è stato quindi snellito, lasciando soltanto le righe riguardanti i comuni siciliani. 

8. Sono stati modificati i valori di alcune righe/colonne in modo da rendere omogenee le informazioni che i dataset hanno in comune

9. Sono stati uniti tramite concatenazione i valori di certe colonne come "Prefisso telefono" e "Telefono" in un'unica colonna.

```python
# PARTE 1 - SET UP  


from rdflib import Graph, URIRef, BNode, Literal, Namespace, Bag
from rdflib.namespace import RDF, OWL, XSD


# 1- creo il grafo
g= Graph()

# - creo URI BASE che verrà utilizzato per creare URI per le risorse.
uribase= "http://www.popsalute.it/resource/"


# 3- definisco namespace per ontology
#pso = Namespace("http://popsalute.it/ontology/")
pso = Namespace("http://www.popsalute.it/ontology/")


# stessa cosa per resource
psr = Namespace("http://www.popsalute.it/resource/")


# 4- collego i prefissi ai rispettivi namespaces
g.bind("pso", pso)
g.bind("psr", psr)


def urify(uribs, resource):
    adapted = resource.replace(" ", "_")
    
    #rimpiazza gli apostrofi \' con "" 
    adapted = adapted.replace("\'", "")
    
    #parse.quote rende safe il link, nel caso in cui avesse caratteri illeciti
    #per lo standard RFC 3986 
    return uribs + urllib.parse.quote(adapted)

print("OK")
```

```python
# PARTE 1.5 - INTERLINKING COMUNI CON DBPEDIA ------------ 

from SPARQLWrapper import SPARQLWrapper2        



"""LA FUNZIONE VIENE CHIAMATA OGNI VOLTA CHE SI AGGIUNGE UN NUOVO COMUNE, ALL'INTERNO
   DELLA FUNZIONE comuniRDF"""


def dbpediaInterlink(res_com, nome_com):
    
    #set dell'endpoint
    sparql = SPARQLWrapper2("http://dbpedia.org/sparql")

    myquery= f"""
             SELECT ?comune ?label
             WHERE {{
                       ?comune dbo:region dbr:Sicily .
                       ?comune rdfs:label ?label .
            
                       FILTER (LANG(?label) = "it")
                       FILTER regex(?label, "{nome_com}", "i") .
         
                    }}
             """
        
        # FILTER regex(?label, "{nome_com}", "i")  --> filtra tramite regex il comune che corrisponde a nome_com
        # "i" vuol dire "case insensitive"
        
    #setquery --> imposta la query che sarà eseguita
    sparql.setQuery(myquery)
     
        
    # Esegue la query e
    # memorizza i risultati (già in JSON) nella var "risultato"
    risultati = sparql.query()
    
    
    #se c'è almeno un risultato
    if(len(risultati.bindings)>0):
        
        b_res= risultati.bindings[0]
        
        # aggiunge a res_com il link dbpedia dello stesso comune
        g.add([res_com, OWL.sameAs, URIRef(b_res["comune"].value)])
        
        #DEBUG
        print(nome_com + " --> " + b_res["comune"].value)
        
              

      
        #DEBUG stampa i bindings (risultati) della query
        #for binding in risultato.bindings:
        # ogni binding è un dictionary (coppia key-value) 
        #print(f"{binding['comune'].value}, {binding['comune'].type}")
        #print(f"{binding['label'].value}, {binding['label'].type}")


print("OK")
```

```python
# PARTE 2 - INSERIMENTO DEI NODI RIGUARDANTI REGIONE, PROVINCIE, COMUNI e POPOLAZIONE NEL GRAFO RDF  


import urllib.parse      


    
def comuniRdf(dspop, dsregion):
    
      # http://www.popsalute.it/resource/Sicilia
    res_region = URIRef(urify(uribase, "Sicilia"))
    
    #region è una risorsa di tipo Regione
    g.add([res_region, RDF.type, URIRef(pso.Regione)])
    
    
    row_sici= dsregion.loc[dsregion["TERRITORIO"] == "Sicilia"]
    
    cod_sici= row_sici["CODICE ISTAT"].iloc[0]
    pop_sici= row_sici["VALORE"].iloc[0] # VALORE è LA POPOLAZIONE TOTALE
    
    
    
    g.add([res_region, pso.haCodiceRegione, Literal(cod_sici, datatype=XSD.string) ])
    g.add([res_region, pso.haPopolazione, Literal(pop_sici, datatype=XSD.nonNegativeInteger)])
   

    
    
    #PROVINCE *****************************************************
    
    # parte da 1 e arriva a 10 (perchè 0 è la sicilia)
    for p in range(1,len(dsregion)):
        
        
        
        # http://www.popsalute.it/resource/nome_provincia
        res_provincia = URIRef(urify(uribase, dsregion["TERRITORIO"][p]+"_provincia"))
        #print(res_provincia)
        
        #provincia è di tipo pso.Provincia
        g.add([res_provincia, RDF.type, URIRef(pso.Provincia)])
        
        #add [provincia p] a regione
        g.add([res_region, pso.haProvincia , res_provincia])
        
        # ADD CODICE ISTAT e POPOLAZIONE A PROVINCIA
        g.add([res_provincia, pso.haCodiceIstat,   Literal(dsregion["CODICE ISTAT"][p], datatype=XSD.string)])
        g.add([res_provincia, pso.haPopolazione, Literal(dsregion["VALORE"][p], datatype=XSD.nonNegativeInteger)])
     
    
#COMUNI *****************************************************************
        
        #inserisce tutti i comuni
        
    for c in range(len(dspop)):

                   
        # http://www.popsalute.it/resource/Comune
        res_comune = URIRef(urify(uribase, dspop["DENOMINAZIONE COMUNE"][c]))
       

         #la res appena creata è di tipo "Comune"
        g.add([res_comune, RDF.type, URIRef(pso.Comune)])
                
                
        g.add([res_comune, pso.haCodiceRegione, Literal(dspop["CODICE REGIONE"][c],datatype=XSD.nonNegativeInteger)])
        g.add([res_comune, pso.haCodiceIstat, Literal(dspop["CODICE COMUNE ISTAT"][c], datatype=XSD.string)])
        g.add([res_comune, pso.haDenominazione, Literal(dspop["DENOMINAZIONE COMUNE"][c], datatype=XSD.string)])
        g.add([res_comune, pso.haRegione, Literal(dspop["REGIONE"][c], datatype=XSD.string)])
        g.add([res_comune, pso.haProvincia, Literal(dspop["PROVINCIA"][c], datatype=XSD.string)])
        g.add([res_comune, pso.haSiglaAutomobilistica, Literal(dspop["SIGLA AUTOMOBILISTICA"][c], datatype=XSD.string)])
        g.add([res_comune, pso.haPopolazione, Literal(dspop["POPOLAZIONE"][c], datatype=XSD.nonNegativeInteger)])

                #aggiungo il comune alla provincia
        g.add([URIRef(urify(uribase, dspop["PROVINCIA"][c]+"_provincia")), pso.haComune, res_comune])    
                
                
                #INTERLINK ************
        dbpediaInterlink(res_comune, dspop["DENOMINAZIONE COMUNE"][c])
                
    return


#comuni e popolazione
popolazione = pd.read_csv(r"C:\Users\ddmon\Desktop\ospedali\3-ready\popolazione.csv", dtype=str, 
                    encoding= "utf8", sep= ",")

regione = pd.read_csv(r"C:\Users\ddmon\Desktop\ospedali\3-ready\popolazione_rp.csv", dtype=str, 
                    encoding= "utf8", sep= ",")
    

# creo grafo rdf comuni, passandogli i dataset comuni+popolazione e regione
comuniRdf(popolazione, regione)


print("OK")
   
```

```python
#CREAZIONE GRAFO RDF STRUTTURE SANITARIE


def sanitRdf(dflenght):
    
    popcom = pd.read_csv(r"C:\Users\ddmon\Desktop\ospedali\3-ready\popolazione.csv", dtype=str, 
                    encoding= "utf8", sep= ",")
    
  
    for i in range(dflenght):
            
        
        
        #Creo URI della risorsa attuale e poi l'aggiungo al grafo, ed il suo tipo (di res) è StrutturaSanitaria 
        st_sanit = URIRef(urify(uribase,dssanit["CODICE STRUTTURA"][i]))
       
        
        
        
       # check per verificare se una struttura è già stata inserita 
        if(st_sanit, RDF.type, URIRef(pso.StrutturaSanitaria)) in g:
            
            
            
            #Quando si entra in questo blocco, è perchè la risorsa è già presente nel grafo,
            # e si deve aggiungere un nuovo reparto: quindi verranno inserite solo le info che riguardano
            # il nuovo reparto da aggiungere (che sono le info non in comune)
                         
            row_prov = popcom.loc[popcom["DENOMINAZIONE COMUNE"] == dssanit["DENOMINAZIONE COMUNE"][i]]
            
                
                #aggiungo alla struttura il comune in cui si trova, con la obj property "isIn"
                #prelevando la provincia dal dataset dei comuni/popolazione
            g.add([st_sanit, pso.isIn, URIRef(urify(uribase,dssanit["DENOMINAZIONE COMUNE"][i]))])             
            
            
   
             #entity reparto
            reparto=  URIRef(urify(uribase, dssanit["CODICE STRUTTURA"][i]+"_" + dssanit["TIPO ASSISTENZA EROGATA"][i]))
           
            
             #specifico che reparto è di tipo: pso.Reparto
            g.add([reparto, RDF.type, URIRef(pso.Reparto)])
                
                        

            g.add([reparto, pso.haTipoAssistenza, Literal(dssanit["TIPO ASSISTENZA EROGATA"][i], datatype= XSD.string)])
            g.add([reparto, pso.haDescrizioneTipoAssistenza, Literal(dssanit["DESCRIZIONE TIPO ASSISTENZA EROGATA"][i], datatype= XSD.string)])      
            g.add([reparto, pso.haAnnoApertura,Literal(dssanit["ANNO APERTURA"][i], datatype= XSD.unsignedInt)])
            g.add([reparto, pso.haAnnoChiusura,Literal(dssanit["ANNO CHIUSURA"][i], datatype= XSD.unsignedInt)])     
            g.add([reparto, pso.haDataAggiornamento,Literal(dssanit["DATA AGGIORNAMENTO"][i], datatype= XSD.string)]) 
            g.add([reparto, pso.haPresenzaApparecchiature, Literal(dssanit["PRESENZA APPARECCHIATURE"][i], datatype= XSD.string)])
            g.add([reparto, pso.haTipoRapportoSSN, Literal(dssanit["TIPO RAPPORTO CON IL S.S.N."][i], datatype= XSD.unsignedInt)])
            
            
             #aggiungo l'entità reparto a res_strutt sanit  
            g.add([st_sanit, pso.haReparto, reparto])
             
            
           
           
            
            
                
        else:
            
            #Se si entra in questo blocco, è perchè la risorsa non è ancora presente nel grafo.
            #Quindi viene creato il nodo padre e vengono aggiunte le info comuni e non comuni, per una stessa struttura.
        
            
            g.add([st_sanit, RDF.type, URIRef(pso.StrutturaSanitaria)])
            
            
            # PROPERTIES(NOT in entity)
  
            g.add([st_sanit, pso.haDenominazione, Literal(dssanit["DENOMINAZIONE STRUTTURA"][i], datatype= XSD.string)])             
            g.add([st_sanit, pso.haCodiceAzienda, Literal(dssanit["CODICE AZIENDA"][i], datatype= XSD.unsignedInt)])
            g.add([st_sanit, pso.haCodiceStruttura, Literal(dssanit["CODICE STRUTTURA"][i], datatype= XSD.nonNegativeInteger)])
            g.add([st_sanit, pso.haPartitaIva, Literal(dssanit["PARTITA IVA"][i], datatype= XSD.string)])        
           
      
        
             #entity reparto
            reparto=  URIRef(urify(uribase, dssanit["CODICE STRUTTURA"][i]+"_" + dssanit["TIPO ASSISTENZA EROGATA"][i]))
            
             #specifico che reparto è di tipo: pso.Reparto
            g.add([reparto, RDF.type, URIRef(pso.Reparto)])
                
           
            g.add([reparto, pso.haTipoAssistenza, Literal(dssanit["TIPO ASSISTENZA EROGATA"][i], datatype= XSD.string)])
            g.add([reparto, pso.haDescrizioneTipoAssistenza, Literal(dssanit["DESCRIZIONE TIPO ASSISTENZA EROGATA"][i], datatype= XSD.string)])      
            g.add([reparto, pso.haAnnoApertura,Literal(dssanit["ANNO APERTURA"][i], datatype= XSD.unsignedInt)])
            g.add([reparto, pso.haAnnoChiusura,Literal(dssanit["ANNO CHIUSURA"][i], datatype= XSD.unsignedInt)])     
            g.add([reparto, pso.haDataAggiornamento,Literal(dssanit["DATA AGGIORNAMENTO"][i], datatype= XSD.string)]) 
            g.add([reparto, pso.haPresenzaApparecchiature, Literal(dssanit["PRESENZA APPARECCHIATURE"][i], datatype= XSD.string)])
            g.add([reparto, pso.haTipoRapportoSSN, Literal(dssanit["TIPO RAPPORTO CON IL S.S.N."][i], datatype= XSD.unsignedInt)])
            
            
               #aggiungo l'entità parto a res_struttura   
            g.add([st_sanit, pso.haReparto, reparto])
            
             
             # V1 - SE LA STRUTTURA HA UN CODICE STRUTTURA DI RICOVERO,
                # ALLORA VIENE RICHIAMATA LA FUNZIONE ricovRDF che aggiunge la struttura di ricovero
                # come figlio della struttura sanitaria
                
                #QUINDI I DATI COME UBICAZIONE/RECAPITO/ECC VENGON AGGIUNTI TRAMITE LA FUNZIONE
                # ricovRDF
                
            if(dssanit["CODICE STRUTTURA DI RICOVERO"][i]!="0"):
                #se true vuol dire che è una struttura di ricovero
                ricovRdf(dsricov, dssanit["CODICE STRUTTURA DI RICOVERO"][i], st_sanit)
                
              
            
            else:
               
            
                g.add([st_sanit, pso.haEmail, Literal(dssanit["E-MAIL"][i], datatype= XSD.string)])
                g.add([st_sanit, pso.haSitoWeb, Literal(dssanit["SITO WEB"][i], datatype= XSD.string)])
                g.add([st_sanit, pso.haTelefono, Literal(dssanit["TELEFONO"][i], datatype= XSD.string)])
                g.add([st_sanit, pso.haNumeroFax, Literal(dssanit["NUMERO FAX"][i], datatype= XSD.string)])

        
       

                g.add([st_sanit, pso.haRegione, Literal(dssanit["REGIONE"][i], datatype= XSD.string)])  
                g.add([st_sanit, pso.haCodiceRegione, Literal(dssanit["CODICE REGIONE"][i], datatype= XSD.unsignedInt)]) 
                g.add([st_sanit, pso.haProvincia, Literal(dssanit["PROVINCIA"].iloc[0], datatype= XSD.string)])
                g.add([st_sanit, pso.haComune, Literal(dssanit["DENOMINAZIONE COMUNE"][i], datatype= XSD.string)])
                g.add([st_sanit, pso.haCodiceIstat, Literal(dssanit["CODICE COMUNE ISTAT"][i], datatype= XSD.string)])
                g.add([st_sanit, pso.haCap, Literal(dssanit["CAP"][i], datatype= XSD.unsignedInt)])
                g.add([st_sanit, pso.haIndirizzo, Literal(dssanit["INDIRIZZO"][i], datatype= XSD.string)])
                
   
    return
    
    
    
    
dssanit= pd.read_csv(r"C:\Users\ddmon\Desktop\ospedali\3-ready\sanit.csv",encoding= "utf8",  dtype=str, sep= ",")
dsricov= pd.read_csv(r"C:\Users\ddmon\Desktop\ospedali\3-ready\ospedali.csv",encoding= "utf8",  dtype=str, sep= ",")
    
    
sanitRdf(len(dssanit))
               
print("OK")      
```
